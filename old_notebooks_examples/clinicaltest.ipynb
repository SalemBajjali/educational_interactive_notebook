```python
def create_spdi_df(dataset):

    # Copy dataset
    data = dataset.copy()

    # Converting from 1 base to 0 base indexing 
    data['Position_start'] = (data['Position_start'] - 1)

    # Subset data 
    subset_data = data[['RefSeq','Position_start','Ref','Variant']]

    # Concatenate columns
    subset_data['spdi_expr'] = subset_data['RefSeq'].str.cat(subset_data[['Position_start', 'Ref','Variant']].astype(str), sep=":")

    return subset_data
```
