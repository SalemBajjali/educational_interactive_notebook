```python
import pandas as pd
from src.extra_translate import extra
from src.variation_services_rest_db import VariationServicesRESTDataProxy
convert = extra()
converting_hgvs = VariationServicesRESTDataProxy()
```

```python
from ga4gh.vrs.dataproxy import SeqRepoRESTDataProxy
from ga4gh.vrs.extras.translator import Translator
from ga4gh.vrs.extras.variation_normalizer_rest_dp import VariationNormalizerRESTDataProxy
vnorm = VariationNormalizerRESTDataProxy()

seqrepo_rest_service_url = "https://services.genomicmedlab.org/seqrepo"
dp = SeqRepoRESTDataProxy(base_url=seqrepo_rest_service_url)
tlr = Translator(dp)
```

```python
df = pd.read_csv('/Users/M278428/Documents/educational_notebooks/data/dbVar_structural_variants.csv',dtype = str)
df
```

```python
def from_hgvs_to_spdi(expression):
    try: 
        # Converting a allele in SPDI syntax to the right-shifted HGVS notation
        return converting_hgvs.hgvs_to_spdi(expression)
    except Exception as e: 
        # returns error produce by NCBI API 
        return 'Error in expression {}'.format(e)     
```

```python
convert_to_spdi = []
correct_spdi_expression= []
spdi_expressions = []

index_of_incorrect_spdi_expr= []
index_of_correct_spdi_expr= []

for sample in df['description']:
    convert_to_spdi.append(from_hgvs_to_spdi(sample)) 

for idx,spdi in enumerate(convert_to_spdi):
    index_of_incorrect_spdi_expr.append(idx)
    if isinstance(spdi,list):
        index_of_correct_spdi_expr.append(idx)
        correct_spdi_expression.append(spdi)

for sublist in correct_spdi_expression:
    spdi_expressions.extend(sublist)
```

```python
spdi_expressions.append('hahahahha')
spdi_expressions
```

```python
vrs_alleles = []
hgvs_expressions = []

for expression in spdi_expressions:
    try:
        trans = tlr.translate_from(expression, 'spdi')
        vrs_alleles.append(trans)
    except Exception as e:
        # returns error produce by translate_from method
        vrs_alleles.append('Error in expression {}'.format(e))

# Removing string values 
vrs_alleles = [x for x in vrs_alleles if not isinstance(x, str)]

hgvs_expressions = []
for allele in vrs_alleles:  
    try:  
        # translating the vrs objects to hgvs expressions
        hgvs_expressions.append(vnorm.to_hgvs(allele, 'refseq'))
    except Exception as e:
        # returns error produce by to_hgvs method
        hgvs_expressions.append('Error in expression {}'.format(e))
```

```python
hgvs_expressions
```

```python
allForms = {
    'input_description' : df.iloc[index_of_correct_spdi_expr]['description'],
    'ga4gh_id' : [],
    'vrs_allele_object' : [],
    'spdi_expression' : correct_spdi_expression,
    'fullynorm_hgvs' : hgvs_expressions,
    'rightshift_hgvs' : []
    }

vrs_object_list = []

for sample in spdi_expressions:
    allForms['rightshift_hgvs'].append(convert.to_rightshift_hgvs(sample))
    # allForms['fullynorm_hgvs'].append(convert.to_fullynorm_hgvs(sample))
    vrs_object_list.append(convert.to_vrs_allele(sample))  
    
for vrs_objects in vrs_object_list:
    for vrs_id in vrs_objects.keys():
        allForms['ga4gh_id'].append(vrs_objects.keys())
        allForms['vrs_allele_object'].append(vrs_objects[vrs_id])


allFormDF = pd.DataFrame.from_dict(allForms)
allFormDF
```
