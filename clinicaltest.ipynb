```python
import pandas as pd
from src.extra_translate import extra
convert = extra()
```

```python
df = pd.read_csv('~/data/sample_data_1.csv')
df["RefSeq"] = df["RefSeq"].str.replace("\t","")
```

```python
df
```

```python
def create_spdi_df(dataset):

    # Copy dataset
    data = dataset.copy()

    # Converting from 1 base to 0 base indexing 
    data['Position_start'] = (data['Position_start'] - 1)

    # Subset data 
    subset_data = data[['RefSeq','Position_start','Ref','Variant']]

    # Concatenate columns
    subset_data['spdi_expr'] = subset_data['RefSeq'].str.cat(subset_data[['Position_start', 'Ref','Variant']].astype(str), sep=":")

    return subset_data

# Capturing the spdi expressions
subdf = create_spdi_df(df)

# Creating a dictionary that contains all necessary formats. 
allForms = {
    'ga4gh_id' : [],
    'vrs_allele_object' : [],
    'rightshift_hgvs' : [],
    'fullynorm_hgvs' : []
    }

vrs_object_list = []

for sample in subdf['spdi_expr']:
    allForms['rightshift_hgvs' ].append(convert.to_rightshift_hgvs(sample))
    allForms['fullynorm_hgvs'].append(convert.to_fullynorm_hgvs(sample))
    vrs_object_list.append(convert.to_vrs_allele(sample))  
    
for vrs_objects in vrs_object_list:
    for vrs_id in vrs_objects.keys():
        allForms['ga4gh_id'].append(vrs_objects.keys())
        allForms['vrs_allele_object'].append(vrs_objects[vrs_id])
```

```python
allFormDF = pd.DataFrame.from_dict(allForms)
```
